**bore-stroke: -0.055390:** very weak negative linear relationship

**bore-compression-ratio: 0.001263:** extremely weak or negligible linear relationship

**stroke-horsepower: 0.098462:** very weak positive linear relationship

**compression-ratio-horsepower: -0.214514:** weak negative linear relationship

```python
# Engine size as potential predictor variable of price

print(df[['engine-size', 'price']].corr())

sns.regplot(x="engine-size", y="price", data=df)
plt.ylim(0,)
```

There is a positive direct correlation between engine-size and price. So, engine-size seems like a good predictor of price because the regression line is almost a perfect diagonal line.

```python
pearson_coef, p_value = stats.pearsonr(df['engine-size'], df['price'])

print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value)
```

Since the p-value is
 < 0.001, the correlation between engine-size and price is statistically significant and the linear relationship is strong positive correlation (~0.872).

```python
print(df[['highway-mpg', 'price']].corr())

sns.regplot(x="highway-mpg", y="price", data=df)
```

There is a negative relationship between highway-mpg and price and the correlation coefficient says that there is a strong negative linear relationship between highway mpg and price so it can be a potential predictor of price.

```python
pearson_coef, p_value = stats.pearsonr(df['highway-mpg'], df['price'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value)
```

Since the p-value is < 0.001, the correlation between highway-mpg and price is statistically significant and the linear relationship is strong negative correlation (~ -0.704).

```python
print(df[['peak-rpm', 'price']].corr())

sns.regplot(x="peak-rpm", y="price", data=df)
```

The regression line is close to horizontal and the correleation is very weak negative linear relationship. Also, the data points are very scattered and far from the fitted line. So, it shows that peak-rpm is not a good predictor of price.

```python
pearson_coef, p_value = stats.pearsonr(df['peak-rpm'], df['price'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value)
```

Since the p-value is > 0.05, the correlation between highway-mpg and price is not statistically significant and The observed correlation is very weak (~ -0.102), suggesting a minimal linear relationship between the two variables.

```python
print(df[["stroke","price"]].corr())

sns.regplot(x="stroke", y="price", data=df)
```

There is a weak correlation between the variable 'stroke' and 'price.' as such regression will not work well.

```python
sns.boxplot(x="body-style", y="price", data=df)
```

The distributions of price between the different body-style categories have a significant overlap, so body-style would not be a good predictor of price.

```python
sns.boxplot(x="engine-location", y="price", data=df)
```

The distribution of price between these two engine-location categories, front and rear, are distinct enough to take engine-location as a potential good predictor of price.

```python
sns.boxplot(x="drive-wheels", y="price", data=df)
```

The distribution of price between the different drive-wheels categories differs. So, drive-wheels could potentially be a predictor of price.

```python
fig, ax = plt.subplots()
im = ax.pcolor(grouped_pivot, cmap='RdBu')

#label names
row_labels = grouped_pivot.columns.levels[1]
col_labels = grouped_pivot.index

#move ticks and labels to the center
ax.set_xticks(np.arange(grouped_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(grouped_pivot.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```
